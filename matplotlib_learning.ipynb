#### 画条形图

```python
np.random.seed(0)
x = np.arange(5)
y = np.random.randint(-5, 5, 5)
```

```python
fig, axes = plt.subplots(ncols=2)
v_bars = axes[0].bar(x, y, color='red')
h_bars = axes[1].barh(x, y, color='g')
axes[0].axhline(0, color='grey')  # 在y=0加一条线
axes[1].axvline(0, color='grey')  # 在x=0加一条线
```

```python
fig, ax = plt.subplots()
v_bars = ax.bar(x, y, color='lightblue')
ax.axhline(0, color='grey')
for bar, height in zip(v_bars, y):
    if height < 0:
        bar.set(edgecolor='darkred', color='g')
```

```python
x = np.random.randn(100).cumsum() # 计算轴向元素累加和，返回由中间结果组成的数组
y = np.linspace(0, 10, 100)
fig, axes = plt.subplots(ncols=2)
axes[0].fill_between(x, y, color='red')  # 图区域填充
axes[1].plot(x, y)
```

```python
x = np.linspace(0, 10, 200)
y1 = 2 * x + 1
y2 = 3 * x + 1.2
y_mean = 0.5 * x * np.cos(2 * x) + 2.5 * x + 1.1
fig, axes = plt.subplots()
axes.fill_between(x, y1, y2, color='b')
axes.plot(x, y_mean, color='k')
```

```python
mean_values = [1, 2, 3]
variance = [0.2, 0.4, 0.5]
bar_label = ['bar1', 'bar2', 'bar3']

x_pos = list(range(len(bar_label)))
plt.bar(x_pos, mean_values, yerr=variance)
max_y = max(zip(mean_values, variance))
plt.ylim([0, (max_y[0]+max_y[1]) * 1.2])
plt.ylabel('variance_y')
plt.xticks(x_pos, bar_label)
```

#### 某条轴不可见

```python
import math
x = np.random.normal(loc=0.0, scale=1.0, size=300)
width = 0.5
bins = np.arange(math.floor(x.min())-width, math.ceil(x.max())+width, width)

axes = plt.subplot(111)
axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)
axes.tick_params(bottom=True, left=False, right=False, top=False) #刻度是否可见
axes.grid()  # 加网格
axes.hist(x, alpha=0.5, bins=bins)
```

#### 3D图

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
axes = Axes3D(fig)
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(x, y)  # 构造网格

Z = np.sin(np.sqrt(x**2 + y**2))
axes.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
```

#### 3d-散点图

```python
np.random.seed(1)
def randrange(n, v_min, v_max):
    return (v_max -v_min) * np.random.rand(n) + v_min
fig = plt.figure()
axes = fig.add_subplot(111, projection='3d')
n = 100
for c, m, z_low, z_high in [('r', 'o', -50, -25), ('g', '+', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, z_low, z_high)
    axes.scatter(xs, ys, zs, c=c, marker=m)

axes.view_init(60, 0)  # 调整视图角度
```
