例：

| 工资 | 年龄 | 额度 |
| ----- | ----- | ----- |
| 4000 | 25 | 20000 |
| 8000 | 30 | 70000 |
| 5000 | 28 | 35000 |
| 7500 | 33 | 50000 |
| 12000 | 40 | 85000 |

- 数据： 工资和年龄(2个特征)
- 目标： 预测银行贷款额度
- 考虑： 工资和年龄都会影响最终银行贷款额度，则各有多大影响？

则：
$$Y = x_{1}\theta_{1} + x_{2}\theta_{2}\dotsm(1)
(其中Y，x_{1}，x_{2}均已知，求\theta_{1}，\theta_{2})$$

有个问题，该表达式是线性的，不能拟合所有数据点。因此需要找到最合适的一条线(平面)来最好的拟合所有数据点

- 假设$\theta_{1}$是年龄的参数，$\theta_{2}$是工资的参数
- 拟合的平面：$h_{\theta}(x) = \theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2}$，其中$\theta_{0}$为偏置项(使该平面更精确而做一定的微调)
- 把所有数据整合一下：
$$h_{\theta}(x) = \displaystyle\sum\limits_{i=0}^n \theta_{i}x_{i} = \theta^TX$$(其中$\theta$是一个行矩阵，[$\theta_{1}, \theta_{2}, \theta_{3}, \dotsm$]，$X$是一个6x2的矩阵(每一行表示一条数据，每一列表示每个数据的某一特征)，后续对数据的操作都是对矩阵的操作)(心说：矩阵运算都忘完了)

|$x_{0}$ | $x_{1}$(工资) | $x_{2}$(年龄) | Y(额度) |
| :-----: | :-----: | :-----: | :-----: |
| 1 | 4000 | 25 | 20000 |
| 1 | 8000 | 30 | 70000 |
| 1 | 5000 | 28 | 35000 |
| 1 | 7500 | 33 | 50000 |
| 1 | 12000 | 40 | 85000 |

那问题又来了，现在$x_{1}, x_{2}, \theta_{1}, \theta_{2}$都知道了，那拟合的平面就出来了，给一个样本数据就能得出一个额度， 那这个额度和银行给出的额度肯定会有一定误差的。那对于每一个样本就有了：
$$ y^{(i)} = \theta^Tx^{(i)} + \epsilon^{(i)} $$
- 其中$y^{(i)}$表示真实值，$\theta^Tx^{(i)}$表示预测值，$\epsilon^{(i)}$表示误差值。

那，我们的目的当然是预测值和真实值越接近越好啊

---

### 误差
- 误差$\epsilon^{(i)}$是独立分布并且具有相同的分布，且服从均值为0，方差为$\theta^{2}$的高斯分布
    - 独立：张三和李四一起来贷款，他俩没关系
    - 同分布：他俩都来同一家银行贷款
    - 高斯分布：银行可能会多给，也可能会少给，但是绝大多数情况下，这个浮动不会太大，极小情况下浮动会较大， 符合正常情况
        - 高斯分布：$X\sim N(\mu, \sigma^2)：随机变量X的取值x_{i}和其对应的概率值P(X = x_i) 满足正态分布（高斯函数）$
        - 高斯(正态)分布的概率密度函数: $$\large f(x) = \frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$
            - 其中$\mu$表示均值，$\sigma^2$表示方差
        - 概率密度函数: 描述随机变量的输出值落在某个确定的取值点附近的可能性.


- 预测值与误差： $y^{(i)} = \theta^Tx^{(i)} + \epsilon^{(i)}\dotsm\dotsm(1)$
- 由于误差服从高斯分布，误差的概率密度函数为：$$p(\epsilon^{i}) = \frac{1}{\sqrt{2\pi}\sigma}\exp(-\frac{(\epsilon^{(i)})^2}{2\sigma^2})\dotsm\dotsm(2)$$
- 将(1)式带入(2)式得到：$$p(y^{(i)}|x^{(i)};\theta) = \frac{1}{\sqrt{2\pi}\sigma}\exp(-\frac{(y^{(i)}-\theta^{T}x^{(i)})^{2}}{2\sigma^2})$$

### 求解
- 似然函数：是一种关于统计模型参数的函数，表示模型参数中的似然性。概率描述了已知参数时的随机变量的输出结果；似然则用来描述已知随机变量输出结果时，未知参数的可能取值。通俗地讲就是通过样本数据去估计参数取值。
---
在大量样本的情况下，可以对$\theta$做参数估计：

- $\theta$的似然函数：$L(\theta) = \prod\limits_{i=1}^m p(y^{(i)}|x^{(i)};\theta) = \prod\limits_{i=1}^m \frac{1}{\sqrt{2\pi}\sigma}\exp(-\frac{(y^{(i)}-\theta^{T}x^{(i)})^{2}}{2\sigma^2})$
    - 其中$m$表示样本数
    - 解释似然函数：什么样的参数和数据组合后跟真实值越接近越好(即预测值成为真实值的概率值越大越好)
    - 注意：累乘？怎么解？再见，能力有限，不会解
 累乘不会解，那累加会解吗？
 - 对数似然: $l(\theta) = \log L(\theta)$
     - $l(\theta) = \log\prod\limits_{i=1}^m \frac{1}{\sqrt{2\pi}\sigma}\exp(-\frac{(y^{(i)}-\theta^{T}x^{(i)})^{2}}{2\sigma^2})$
     - 即 $l(\theta) = \sum\limits_{i=1}^m \log\frac{1}{\sqrt{2\pi}\sigma}\exp(-\frac{(y^{(i)}-\theta^{T}x^{(i)})^{2}}{2\sigma^2})$
     - 化简得  $l(\theta) = \sum\limits_{i=1}^m \log\frac{1}{\sqrt{2\pi}\sigma} + \sum\limits_{i=1}^m(-\frac{(y^{(i)}-\theta^{T}x^{(i)})^{2}}{2\sigma^2})$
     - 再化简得 $l(\theta) = m\log\frac{1}{\sqrt{2\pi}\sigma} + (-\frac{1}{2\sigma^2})\sum\limits_{i=1}^m {(y^{(i)}-\theta^{T}x^{(i)})^{2}}$
     - $l(\theta) = m\log\frac{1}{\sqrt{2\pi}\sigma} - \frac{1}{2\sigma^2}\sum\limits_{i=1}^m {(y^{(i)}-\theta^{T}x^{(i)})^{2}}$
         - 其中$m\log\frac{1}{\sqrt{2\pi}\sigma}和\frac{1}{2\sigma^2}$均为常数
 
 - 求解目标：让预测值越接近真实值越好， 即让似然函数值越大越好
     - 即 $J(\theta) = \frac{1}{2}\sum\limits_{i=1}^m {(y^{(i)}-\theta^{T}x^{(i)})^{2}}$越小越好

- 最大似然估计
    - 目标函数：$J(\theta) = \frac{1}{2}\sum\limits_{i=1}^m {(y^{(i)}-\theta^{T}x^{(i)})^{2}}$
    - 展开 $J(\theta) = \frac{1}{2}(X\theta - Y)^T(X\theta - Y)$
        - 其中$X, Y$表示矩阵，其中$X=\pmatrix{x^{(1)}\\x^{(2)}\\\vdots\\x^{(m)}}$，$Y=\pmatrix{y^{(1)}\\y^{(2)}\\\vdots\\y^{(m)}}$ 
    - 求偏导:

&emsp;&emsp;&emsp;&emsp; $\large \nabla_\theta J(\theta) = \nabla_\theta \pmatrix{\frac{1}{2}(X\theta - Y)^T(X\theta - Y)}$

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\large = \nabla_\theta \pmatrix{\frac{1}{2}(\theta^TX^T - Y^T)(X\theta - Y)}$

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\large = \nabla_\theta \pmatrix{\frac{1}{2}(\theta^TX^TX\theta - \theta^TX^TY - Y^TX\theta + Y^TY)}$

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\large = \frac{1}{2}(2X^TX\theta - X^TY - (Y^TX)^T)$

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\large = \frac{1}{2}(2X^TX\theta - X^TY - X^TY)$

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\large = X^TX\theta - X^TY$

- 偏导等于0处取极值， 即$\nabla_\theta J(\theta) = 0 = X^TX\theta - X^TY$

&emsp;&emsp;&emsp;&emsp; $X^TX\theta - X^TY = 0$

&emsp;&emsp;&emsp;&emsp; $X^TX\theta = X^TY$

&emsp;&emsp;&emsp;&emsp; $\theta = (X^TX)^{-1}X^TY$
