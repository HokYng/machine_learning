### 关联规则
- 若两个或多个变量的取值之间存在某种规律性就称为关联
- 关联规则是在寻找同一个时间中出现的不同项的相关性

| $T_{7}$ | 1 | 0 | 1 | 0 | 0 |

| $T_{9}$ | 1 | 1 | 1 | 0 | 0 |

- 一个样本称为一个"事务" (如 上表有9个事务/样本)
- 每个事务由多个属性来确定， 这里的属性称为"项"
- 多个项组成的集合称为"项集"

#### 由k各项构成的集合
- {牛奶}、{啤酒} 都是1项集。购物车里只有一样物品，换句话说这个样本只有一个属性
- {牛奶, 果冻} 是二项集。购物车里有两样物品，换句话说这个样本有两个属性
- {牛奶，啤酒，面包}是三项集

#### 关联规则X==>Y:
- X和Y都是项集
- X称为前项
- Y称为后项

#### 事务仅包含其涉及到的项目，而不包含项目的具体信息
- 在超市的关联规则挖掘问题中事务是顾客一次购物所购买的物品，但事务中并不包含这些商品的具体信息，如价格，数量等

#### 支持度: 一个项集或者规则在所有事务中出现的频率，$\sigma(X)$: 表示项集X的支持度计数(频数)
- 项集X的支持度: $s(X) = \sigma(X) / N$
- 规则X==>Y表示项集X对项集Y的支持度，也就是项集X和项集Y同时出现的概率
- 某天共有100个顾客到商场购买物品，其中有30个顾客同时购买了啤酒和尿布，那么上述的关联规则支持度就是：s({啤酒，尿布}) = 30/100 = 30%

#### 置信度(confidence): 确定 Y 在包含 X的事务中 出现的频繁程度。$c(X \Rightarrow Y) = \sigma(X \bigcup Y) / \sigma(X)$
- $p(Y|X) = p(XY) / p(X)$
- 置信度反应了关联规则的可信度--购买了项集X中的物品的顾客同时也购买了项集Y中商品的可能性有多大
- 购买薯片的顾客中有50%的人购买了可乐，则置信度为50%。

#### (X，Y) ==> Z：
- 支持度：交易中包含{X，Y，Z}的可能性
- 置信度：包含{X，Y}的交易中也包含Z的概率

例：

| 交易ID | 购买的商品 |
| :----: | :----: |
| $T_{1}$ | A，B，C |
| $T_{2}$ | A，C |
| $T_{3}$ | A，D |
| $T_{4}$ | B，E，F |

- A==>C：

    - p(AC) = 2 / 4 = 50%    ---->AC同时出现的是$T_{1}$，$T_{2}$
    - p(C|A) = 2 / 3 = 66.6%    ---->出现A的是$T_{1}$，$T_{2}$，$T_{3}$；在出现A的情况下出现C的是$T_{1}$，$T_{2}$

- C==>A：

    - p(CA) = 2 / 4 = 50%  ---->CA同时出现的是$T_{1}$，$T_{2}$
    - p(A|C) = 2 / 2 = 100%  ----> 出现C的是$T_{1}$，$T_{2}$；在出现C的情况下出现A的是$T_{1}$，$T_{2}$

- (A，C) ==> B：
    - p(ACB) = 1 / 4 = 25%   ---->ACB同时出现的是$T_{1}$
    - p(B|AC) = 1 / 2 = 50%   ---->AC同时出现的是$T_{1}$，$T_{2}$，在出现AC的情况下出现B的是$T_{1}$

若关联规则X==>Y的支持度和置信度分别大于或等于用户指定的最小支持度min(support)和min(confidence)，则称关联规则X==>Y为强关联，反之称为弱关联

#### 提升度(lift)：项集A的出现对项集B的出现概率发生了多大的变化
- lift(A==>B) = confidence(A==>B) / support(B) = p(B|A) / p(B)
- 现在有1000个消费者，有500人购买了茶叶，其中有450人同时购买了咖啡，另50人没有；另外没有购买茶叶的500人，同样有450购买了咖啡
    - 购买了咖啡的概率
        - support(咖啡) = (450+450)/1000 = 90%
        - confidence(茶叶==>咖啡) = 450/500 = 90%
    - 由此可能会认为喜欢喝茶的人往往喜欢喝咖啡。但另外没有购买茶叶的500人，同样有450购买了咖啡，同样置信度也为90%
        - confidence(没买茶叶==>咖啡) = 450/500 = 90%
    - 由此，得到不爱喝茶的也爱喝咖啡。这样看来，其实是否购买咖啡，与有没有购买茶叶并没有关联，两者是相互独立的，其提升度为：
        - lift(茶叶==>咖啡) = confidence(茶叶==>咖啡) / support(咖啡) = 90% / 90% = 1
由此可见，lift正是弥补了confidence的这一缺陷，if lift=1，X与Y独立，X对Y出现的可能性没有提升作用，其值越大(lift>1)，则表明X对Y的提升程度越大，也表明关联性越强
